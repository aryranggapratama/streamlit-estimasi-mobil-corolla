# data understanding

Konteks: mendeskripsikan 1436 mobil corolla dan 9 properti

Isi: 

1.Age : Age adalah usia pada mobil tersebut, berapa lama mobil tersebut di gunakan atau dipakai dari produksi sampai saat ini digunakan

2.KM : KM adalah kilometer atau jarak tempuh mobil tersebut, untuk mengetahui sudah berapa jauh jarak yang di tempuh mobil tersebut

3.Fuel Type : Fuel Type adalah jenis bahan bakar yang digunakan mobil tersebut

4.HP : HP atau HorsePower adalah tenaga yang di keluar kan mobil tersebut

5.Automatic : Automatic adalah transmisi mobil tersebut

6.CC : CC atau Cubical Centimeter adalah Volume ruang silender pada mesin

7.Quarterly_Tax : Quarterly_Tax adalah pajak pada mobil tersebut

8.Weight : Weight adalah berat pada mobil tersebut

9.Price : Price adalah harga mobil

# modeling data

# seleksi fitur

```python
features = ['Age','KM','HP','CC','Quarterly_Tax','Weight']
x = df[features]
y = df['Price']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

# evaluation data

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# membuat inputan model regresi linier

```python
#Age=25.00, KM=60000.00, HP=100.00, CC=100.00, Quarterly_Tax=150.00, Weight=1000.00
input_data = np.array([[25.00,60000.00,100.00,100.00,150.00,1000.00]])

prediction = lr.predict(input_data)
print('Estimasi harga mobil corolla tersebut adalah :', prediction)
```
